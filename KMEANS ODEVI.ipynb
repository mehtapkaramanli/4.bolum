```python
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans
```

```python
heart = pd.read_csv("heart.csv")
```

```python
heart
scaler = StandardScaler()
heart_veri = scaler.fit_transform(heart)
```

```python
k_ortalama = KMeans(n_clusters=2, random_state=123)
%timeit k_ortalama.fit(heart_veri)
y_pred = k_ortalama.predict(heart_veri)

# Plot the solution.
plt.scatter(heart_veri[:,0], heart_veri[:,1], c=y_pred)
plt.show()
```
